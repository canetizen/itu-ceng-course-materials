BLG 374E Rabin Karp Algorithm Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("kmp.csv")
```

```python
# Overflow değeri 0 olan verileri seç
df_overflow_0 = df[df['Overflow'] == 0]

# Her bir hash türü için verileri grupla
grouped = df_overflow_0.groupby('Function_Name')

# Her bir hash fonksiyonu için çizimler yap
for hash_type, group in grouped:
    plt.figure(figsize=(8, 5))  # Yeni bir figür oluştur
    
    # Çizgi grafiğini çiz
    plt.plot(group['Modulus'], group['Collision'], marker='o')
    plt.xlabel('Modulus')
    plt.ylabel('Collision')
    plt.title(hash_type + ' - Collision vs Modulus (Overflow = 0)')
    plt.xscale('log')  # X ekseninde logaritmik ölçek kullan
    plt.grid(True)
    plt.show()
```

```python
# Overflow değeri 0 olan verileri seç
df_overflow_0 = df[df['Overflow'] == 0]

# Her bir hash türü için verileri grupla
grouped = df_overflow_0.groupby('Function_Name')

# Her bir hash fonksiyonu için çizimler yap
for hash_type, group in grouped:
    plt.figure(figsize=(8, 5))  # Yeni bir figür oluştur
    
    # Başarı yüzdesini hesapla
    total_data_points = len(group)
    success_rate = (total_data_points - group['Collision']) / total_data_points * 100
    
    # Çizgi grafiğini çiz
    plt.plot(group['Modulus'], success_rate, marker='o')
    plt.xlabel('Modulus')
    plt.ylabel('Success Rate (%)')
    plt.title(hash_type + ' - Hash Function Success Rate vs Modulus (Overflow = 0)')
    plt.xscale('log')  # X ekseninde logaritmik ölçek kullan
    plt.grid(True)
    plt.show()
```

```python
# Overflow değeri 0 olan verileri seç
df_overflow_0 = df[df['Overflow'] == 0]

# Her bir hash türü için verileri grupla
grouped = df_overflow_0.groupby('Function_Name')

# Yeni bir figür oluştur
plt.figure(figsize=(10, 6))

# Her bir hash türü için çizgi grafiğini çiz
for hash_type, group in grouped:
    # Başarı yüzdesini hesapla
    total_data_points = len(group)
    success_rate = (total_data_points - group['Collision']) / total_data_points * 100
    
    # Çizgi grafiğini çiz
    plt.plot(group['Modulus'], success_rate, marker='o', label=hash_type)

# Grafik özellikleri
plt.xlabel('Modulus')
plt.ylabel('Success Rate (%)')
plt.title('Hash Function Success Rate vs Modulus (Overflow = 0)')
plt.xscale('log')  # X ekseninde logaritmik ölçek kullan
plt.legend()
plt.grid(True)

# Grafik göster
plt.show()
```

```python
# Overflow değeri 0 olan verileri seç
df_overflow_0 = df[df['Overflow'] == 1]

# Her bir hash türü için verileri grupla
grouped = df_overflow_0.groupby('Function_Name')

# Yeni bir figür oluştur
plt.figure(figsize=(10, 6))

# Her bir hash türü için çizgi grafiğini çiz
for hash_type, group in grouped:
    # Başarı yüzdesini hesapla
    total_data_points = len(group)
    success_rate = (total_data_points - group['Collision']) / total_data_points * 100
    
    # Çizgi grafiğini çiz
    plt.plot(group['Modulus'], success_rate, marker='o', label=hash_type)

# Grafik özellikleri
plt.xlabel('Modulus')
plt.ylabel('Success Rate (%)')
plt.title('Hash Function Success Rate vs Modulus (Overflow = 1)')
plt.xscale('log')  # X ekseninde logaritmik ölçek kullan
plt.legend()
plt.grid(True)

# Grafik göster
plt.show()
```

```python
data = pd.read_csv("time_results.csv")
data_nobrute = pd.read_csv("nobrute.csv")
```

```python
# Her bir algoritmayı metin uzunluğuna göre grupla
grouped_data = data.groupby(["Text_Length", "Function_Name"])["Time"].mean().unstack()

# Çizgi grafiği oluştur
plt.figure(figsize=(12, 8))

# Brute Force için çizgi grafiği
plt.plot(grouped_data.index, grouped_data["Brute Force"], label="Brute Force", marker='o')

# KMP için çizgi grafiği
plt.plot(grouped_data.index, grouped_data["KMP"], label="KMP", marker='o')

# Rabin-Karp için çizgi grafiği
plt.plot(grouped_data.index, grouped_data["Rabin-Karp"], label="Rabin-Karp", marker='o')

# Eksen etiketleri ve başlık
plt.xlabel("Text Length")
plt.ylabel("Running Time")
plt.title("Average Running Times of Algorithms Based on Text Length")
plt.legend()  # Göstergeyi ekle

# Grafiği göster
plt.grid(True)
plt.show()
```
